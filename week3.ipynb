```
import cv2 as cv 
from google.colab.patches import cv2_imshow 
import numpy as np
print(' Mengubah tingkat kecerahan citra ')
print('----------------------------------')
try:
  brightness = int(input('Masukkan nilai kecerahan '))
except ValueError:
    print('Error, not a number')

original =  cv.imread('/content/drive/MyDrive/PCVK/P3/female.tiff')
brightness_image = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
       brightness_image[y,x,c] = np.clip(original[y,x,c] + brightness, 0, 255)

final_frame = cv.hconcat((original, brightness_image))
cv2_imshow(final_frame)
```

```
print(' Mengubah kontras dan tingkat kecerahan citra ')
print('----------------------------------')
try:
  brightness = int(input('Masukkan tingkat kecerahan [0 - 255]: '))
  contrast = float(input('Masukkan kontras [1.0 - 3.0]: '))
except ValueError:
    print('Error, not a number')

image_bgcon = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
       image_bgcon[y,x,c] = np.clip(contrast*original[y,x,c] + brightness, 0, 255)

final_frame = cv.hconcat((original, image_bgcon))
cv2_imshow(final_frame)
```

```
print(' Mengubah tingkat kecerahan citra dengan transform log ')
print('----------------------------------')
try:
  brightness = int(input('Masukkan tingkat kecerahan [0 - 255]: '))
except ValueError:
    print('Error, not a number')

image_bgcon = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
       image_bgcon[y,x,c] = np.clip(brightness*(np.log(1+original[y,x,c])), 0, 255)

final_frame = cv.hconcat((original, image_bgcon))
cv2_imshow(final_frame)
```

```
#AVERAGING
import matplotlib.pyplot as plt
image_01 =  cv.imread('/content/drive/MyDrive/PCVK/P3/mandrill.tiff')
image_01 =  cv.cvtColor(image_01,cv.COLOR_BGR2RGB) 
image_02 =  cv.imread('/content/drive/MyDrive/PCVK/P3/mandrill.tiff')
image_02 =  cv.cvtColor(image_02,cv.COLOR_BGR2RGB) 

red = image_01[:,:,0]
green = image_01[:,:,1]
blue = image_01[:,:,2]

avg= (red/3)+(green/3)+(blue/3)
image_02[:,:,0] = avg
image_02[:,:,1] = avg
image_02[:,:,2] = avg

fig, axs = plt.subplots(1, 2) 
axs[0].imshow(image_01)
axs[1].imshow(image_02)
```

```
#LIGHRNESS
image_01 =  cv.imread('/content/drive/MyDrive/PCVK/P3/mandrill.tiff')
image_01 =  cv.cvtColor(image_01,cv.COLOR_BGR2RGB) 
image_03 =  cv.imread('/content/drive/MyDrive/PCVK/P3/mandrill.tiff')
image_03 =  cv.cvtColor(image_02,cv.COLOR_BGR2RGB)

lightness = (np.maximum(red,green,blue)/2) + (np.minimum(red,green,blue)/2)
image_03[:,:,0]=lightness
image_03[:,:,1]=lightness
image_03[:,:,2]=lightness

fig, axs = plt.subplots(1, 2) 
axs[0].imshow(image_01)
axs[1].imshow(image_03)
```

```
#LUMINANCE
image_01 =  cv.imread('/content/drive/MyDrive/PCVK/P3/mandrill.tiff')
image_01 =  cv.cvtColor(image_01,cv.COLOR_BGR2RGB) 
image_03 =  cv.imread('/content/drive/MyDrive/PCVK/P3/mandrill.tiff')
image_03 =  cv.cvtColor(image_02,cv.COLOR_BGR2RGB)

luminance = (0.21*red) + (0.72*green)+ (0.07*blue)
image_03[:,:,0]=luminance
image_03[:,:,1]=luminance
image_03[:,:,2]=luminance

fig, axs = plt.subplots(1, 2) 
axs[0].imshow(image_01)
axs[1].imshow(image_03)
```
